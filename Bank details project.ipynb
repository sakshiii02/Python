```python
dict1={}
account=1000
balance=0
dict2={}
def bankmanagement():
    while True:
        print('='*100)
        print('*'*40,"Welcome to GTM Bank",'*'*40)
        print('='*100)
        print("Enter 1 to create an Account")
        print("Enter 2 to withdraw money from your Account")
        print("Enter 3 to deposit money in your Account")
        print("Enter 4 to view your personal details")
        print("Enter 5 to quit transactions")
        print('='*100)
        option=(int(input("Enter your choice: ")))
        print('='*100)
        if option==1:
            create_account()
        elif option==2:
            withdraw_money()
        elif option==3:
            deposit_money()
        elif option==4:
            check_details()
        elif option==5:
            print('-'*100)
            print('*'*25,"Thanks you!!!!    Visit Again",'*'*25)
            print('-'*100)
            print('='*100)
            break
        else:
            print("Invalid option")
```

```python
def create_account():
    print("You have requested to create an account in GTM bank")
    print('-'*100)
    name=(input("Please Enter your Name: "))
    mobile=(int(input("Please enter your Mobile No: ")))
    Pan=(input("Please Enter your pan card No: "))
    Aadhar=(input("Please Enter your Aadhar card No: "))
    Address=(input("Please Enter your Full Address: "))
    global account
    account+=1
    global balance
    accamt=(int(input("Enter Amount for opening account: ")))
    balance=0+accamt
    while True:
        print('-'*100)
        pin1=int(input("Set a pin for your account: "))
        pin2=int(input("Please confirm your pin again: "))
        print('-'*100)
        if(pin1==pin2):
            print("Pin Setup was successful")
            print('-'*100)
            pin=pin2
            break
        else:
            print("Please check and enter pin properly")
    print("1 to create Savings Account")
    print("2 to create Current Account")
    print('-'*100)
    TypeAcc=(int(input("Enter your choice: ")))
    print('-'*100)
    if TypeAcc==1:
        print("You have created a saving account ")
        Acctype='Saving Account'
        print('-'*100)
    elif TypeAcc==2:
        print("You have created a current account ")
        Acctype='Current Account'
        print('-'*100)
    else:
        print("Invalid Choice")
        print('-'*100)
    Dict1={'Name':name,'Account No':account,'Balance':balance,'Mobile':mobile,'Pan':Pan,'Aadhar':Aadhar,'Address':Address,'Acctype':Acctype,'Pin':pin}
    print('='*100)
    for i in Dict1:
        print(i,' = ',Dict1[i])
    print('='*100)
    print("Account Created Successfully!!!!")
    print('='*100)
    dict1.update({account:[pin,name,balance]})
    print('-'*100)
    print("Please save this details for further use")
    print('-'*100)
    print(f"Mr.{name} you are welcome in our GTM Bank family")
    print("GTM Bank - Graciously Trusted with your Money")
    print('='*100)
    dict2.update({account:list(Dict1.values())})
```

```python
def withdraw_money():
    print("You have choosen to withdraw money")
    print('-'*100)
    account=(int(input("Enter your Account Number:")))
    print('-'*100)
    if (account in dict1.keys()):
        pin=int(input("Enter your pin NO: "))
        print('-'*100)
        if(pin in dict1[account]):
            print(f"welcome Mr.{dict1[account][1]}")
            Amount=int(input("Enter the Amount you want to withdraw: "))
            print('-'*100)
            global balance
            if(balance>=Amount):
                balance=balance-Amount
                print(f"Here is your Rs.{Amount}")
                print(f"Your Remaining Amount is Rs.{balance}")
                print('='*100)
            else:
                print("You don't have sufficient Money")
                print(f"You only have Rs.{balance} left in your Account ")
                print('='*100)
            dict1[account][2]=balance
        else:
            print("Incorrect pin please Enter pin Again")
            print('='*100)
    else:
        print(account)
        print("Incorrect Account Number")
        print('='*100)
```

```python
def deposit_money():
    print("You have choosen to deposit money")
    print('-'*100)
    account=(int(input("Enter your Account Number:")))
    print('-'*100)
    if (account in dict1.keys()):
        pin=int(input("Enter your pin NO: "))
        print('-'*100)
        if(pin in dict1[account]):
            print(f"welcome Mr.{dict1[account][1]}")
            Amount=int(input("Enter the Amount you want to deposit: "))
            print('-'*100)
            global balance
            balance=balance+Amount
            print(f"Thanks for depositing Rs.{Amount}")

            print(f"Your Account now has Rs.{balance}")
            print('='*100)
           
            dict1[account][2]=balance
        else:
            print("Incorrect pin please Enter pin Again")
            print('='*100)
    else:
        print(account)
        print("Incorrect Account Number")
        print('='*100)
```

```python
def check_details():
    print("You have choosen to check your details")
    print('-'*100)
    account=(int(input("Enter your Account Number:")))
    print('-'*100)
    if (account in dict1.keys()):
        pin=int(input("Enter your pin NO: "))
        if(pin in dict1[account]):
            print('-'*100)
            print(f"welcome Mr.{dict1[account][1]}")
            print('='*100)
            print("Your Details are:")
            print('-'*100)
            global dict2
            print("Account holders name:    ",dict2[account][0])
            print("Account Number:          ",dict2[account][1])
            print("Amount is Rs.            ",dict2[account][2])
            print("Mobile Number:           ",dict2[account][3])
            print("PAN Card Number:         ",dict2[account][4])
            print("Aadhar Card Number:      ",dict2[account][5])
            print("Account holders Address: ",dict2[account][6])
            print("Type of Account:         ",dict2[account][7])
            print("PIN:                     ",dict2[account][8])
            print('='*100)
        else:
            print("Incorrect pin please Enter pin Again")
            print('='*100)
    else:
        print(account)
        print("Incorrect Account Number")
        print('='*100)
```
